```python
def computePercentageReceived(totalRejectedQty, totalQuantityRecived, orderedQuantity):
    percentageReceived = 0
    if (totalQuantityRecived > 0.0):
        percentageReceived = ((totalQuantityRecived - totalRejectedQty)/totalQuantityRecived)*100
        percentageReceived = round(percentageReceived, 2)  
    
    '''  
    if (totalQuantityRecived < orderedQuantity):
        percentageReceived = ((totalQuantityRecived - totalRejectedQty)/orderedQuantity) * 100
    else:
        percentageReceived = ((orderedQuantity - totalRejectedQty)/totalQuantityRecived) * 100
    
    quality = 3
    if (percentageReceived == 100):
        quality = 3
    elif (percentageReceived > 80 and percentageReceived < 100):
        quality = 2
    else:
        quality = 1
    return quality
    '''
    return percentageReceived
```

```python
def computeReceiptDiffDays(orderedDateTime, receiptDateTime, needByDateTime):
    diffDays = 0
    
    if (str(needByDateTime) != "1970-01-01 00:00:00"):
        diffDays = (needByDateTime - receiptDateTime).days
    '''
    timeliness = 3
    if (diffDays >= 30):
        timeliness = 1
    elif(diffDays > 10 and diffDays < 30):
        timeliness = 2
    else:
        timeliness = 3
    return timeliness
    '''
    return diffDays
```

```python
def computePOQuality(percentageReceived):
    poQuality = 1.0
    if(percentageReceived >= 0.0 and percentageReceived <= 10.0):
        poQuality = 10.0
    elif (percentageReceived > 10.0 and percentageReceived <= 20.0):
        poQuality = 9.0
    elif (percentageReceived > 20.0 and percentageReceived <= 30.0):
        poQuality = 8.0
    elif (percentageReceived > 30.0 and percentageReceived <= 40.0):
        poQuality = 7.0
    elif (percentageReceived > 40.0 and percentageReceived <= 50.0):
        poQuality = 6.0
    elif (percentageReceived > 50.0 and percentageReceived <= 60.0):
        poQuality = 5.0
    elif (percentageReceived > 60.0 and percentageReceived <= 70.0):
        poQuality = 4.0
    elif (percentageReceived > 70.0 and percentageReceived <= 85.0):
        poQuality = 3.0
    elif (percentageReceived > 85.0 and percentageReceived <= 95.0):
        poQuality = 2.0
    else:
        poQuality = 1.0
    return poQuality
```

```python
def computePOTimeliness(receiptDiffDays):
    poTimeliness = 1.0
    if(receiptDiffDays >= -28.0 and receiptDiffDays <= -25):
        poTimeliness = 1.0
    elif (receiptDiffDays > -25 and receiptDiffDays <= -22.0):
        poTimeliness = 2.0
    elif (receiptDiffDays > -22.0 and receiptDiffDays <= -19.0):
        poTimeliness = 3.0
    elif (receiptDiffDays > -19.0 and receiptDiffDays <= -16.0):
        poTimeliness = 4.0
    elif (receiptDiffDays > -16.0 and receiptDiffDays <= -13.0):
        poTimeliness = 5.0
    elif (receiptDiffDays > -13.0 and receiptDiffDays <= -10.0):
        poTimeliness = 6.0
    elif (receiptDiffDays > -10.0 and receiptDiffDays <= -7.0):
        poTimeliness = 7.0
    elif (receiptDiffDays > -7.0 and receiptDiffDays <= -4.0):
        poTimeliness = 8.0
    elif (receiptDiffDays > -4.0 and receiptDiffDays < -1.0):
        poTimeliness = 9.0
    else:
        poTimeliness = 10.0
    return poTimeliness
```

```python
'''
    Compute supplier characteristics at PO/Item/Supplier level
'''
row = 0
supplierCharacteristicsDf = pd.DataFrame(
    columns=['POId', 'SupplierId', 'SupplierName', 'CommodityId', 'CommodityName', 'OrderedQuantity', 'TotalQuantityReceived', 'TotalQuantityRejected', 'Price', 
             'OrderedDate', 'NeedByDate', 'ReceivedDate', 'PercentageReceived', 'DiffDays', 'POQuality', 'POTimeliness', 'HistQuality', 'HistTimeliness'])

orderReceipts = pd.read_csv('./output/order_receipts.csv')
allPOs = orderReceipts['POId'].unique()

# compute supplier characteristics across all PO/Item level
for i in range(len(allPOs)):
    poID = allPOs[i];
    print("POID = "+poID)
    poReceipts = orderReceipts.query('POId == "'+ poID + '"')
    supplierId = poReceipts['SupplierId'].unique()[0]
    print("SupplierID = "+str(supplierId))
    supplierName = poReceipts['SupplierName'].unique()[0]
    print("SupplierName = "+ str(supplierName))
    poReceiptsAllCommodity = poReceipts['CommodityId'].unique()
    
    # All items of a PO/Supplier
    for j in range(len(poReceiptsAllCommodity)):
        commodityId = poReceiptsAllCommodity[j]
        print("commodityId = " + str(commodityId))
        
        queryStr = 'POId == "' + poID + '" & SupplierId == "' + str(supplierId) + '" & CommodityId == ' + str(commodityId)
        print(queryStr)
        selectedPOReceipts = poReceipts.query(queryStr)
        selectedPOReceipts_last = selectedPOReceipts.iloc[-1]
        commodityName = selectedPOReceipts_last['CommodityName']
        print("CommodityName = " + str(commodityName))
        
        orderedQuantity =  selectedPOReceipts_last['Orderd_Quantity']
        print("Ordered Quantity = " +str(orderedQuantity))
        amount = float(selectedPOReceipts_last['Amount'].replace(',',''))
        print("Amount = "+str(amount))
        price = amount/orderedQuantity
        print("Price = "+str(price))
    
        # Compute item quality at PO/Receipt level
        totalRejectedQty = selectedPOReceipts_last['TotalNumberRejected']
        print("TotalRejectedQty = " +str(totalRejectedQty))
        totalQuantityRecived = selectedPOReceipts_last['TotalQuantityReceived']
        print("TotalQtyReceived = " +str(totalQuantityRecived))
        
        percentageReceived = computePercentageReceived(totalRejectedQty, totalQuantityRecived, orderedQuantity)
        print("PercentageReceived=" + str(percentageReceived))
        poQuality = computePOQuality(percentageReceived)
        print("POQuality=" + str(poQuality))
        
         
        # Compute item timeliness at PO/Receipts level
        orderedDate = selectedPOReceipts_last['OrderedDate']
        orderedDateTime = datetime.strptime(orderedDate, "%m/%d/%Y %I:%M %p")
        receiptDate = selectedPOReceipts_last['ReceiptDate']
        receiptDateTime = datetime.strptime(receiptDate, "%m/%d/%Y %I:%M %p")
        needByDate = selectedPOReceipts_last['LineNeedByDate']
        needByDateTime = datetime.strptime(needByDate, "%m/%d/%Y %I:%M %p")
        print("NeedByDateTime = " + str(needByDateTime))
        
        receiptDiffDays = computeReceiptDiffDays(orderedDateTime, receiptDateTime, needByDateTime)
        print("ReceiptDiffDays = "+ str(receiptDiffDays))
        poTimeliness = computePOTimeliness(receiptDiffDays)
        print("poTimeliness = "+ str(poTimeliness))
        
        
        supplierCharacteristicsDf.loc[row] = [poID, supplierId, supplierName, commodityId, commodityName, orderedQuantity, totalQuantityRecived, totalRejectedQty, price,
                                              orderedDate, needByDate, receiptDate, percentageReceived, receiptDiffDays, poQuality, poTimeliness, 0.0, 0.0]
        row = row + 1 
    print("-------------------------------------------------------------------------------------------------------")

supplierCharacteristicsDf
```

```python
def computeHistQuality(percentageReceivedMin, percentageReceivedMax, percentageReceivedMean):
    print("Hist PercentageReceived ="+ str(percentageReceivedMin) +":"+str(percentageReceivedMax)+":"+str(percentageReceivedMean))
    histQuality = 0.0
    percentageReceivedScaled = 0.0
    if(percentageReceivedMax > percentageReceivedMin):
        percentageReceivedScaled = ((percentageReceivedMean - percentageReceivedMin)/(percentageReceivedMax - percentageReceivedMin))*100
        print(percentageReceivedScaled)
    
    if(percentageReceivedScaled >= 0.0 and percentageReceivedScaled <= 10.0):
        histQuality = 10.0
    elif (percentageReceivedScaled > 10.0 and percentageReceivedScaled <= 20.0):
        histQuality = 9.0
    elif (percentageReceivedScaled > 20.0 and percentageReceivedScaled <= 30.0):
        histQuality = 8.0
    elif (percentageReceivedScaled > 30.0 and percentageReceivedScaled <= 40.0):
        histQuality = 7.0
    elif (percentageReceivedScaled > 40.0 and percentageReceivedScaled <= 50.0):
        histQuality = 6.0
    elif (percentageReceivedScaled > 50.0 and percentageReceivedScaled <= 60.0):
        histQuality = 5.0
    elif (percentageReceivedScaled > 60.0 and percentageReceivedScaled <= 70.0):
        histQuality = 4.0
    elif (percentageReceivedScaled > 70.0 and percentageReceivedScaled <= 85.0):
        histQuality = 3.0
    elif (percentageReceivedScaled > 85.0 and percentageReceivedScaled <= 95.0):
        histQuality = 2.0
    else:
        histQuality = 1.0

    return histQuality
```

```python
def computeHistTimeliness(diffDaysMin, diffDaysMax, diffDaysMean):
    print("Hist Timeliness ="+ str(diffDaysMin) +":"+str(diffDaysMax))
    histTimeliness = 0.0
    
    if(diffDaysMin >= -28.0 and diffDaysMin <= -25):
        histTimeliness = 1.0
    elif (diffDaysMin > -25 and diffDaysMin <= -22.0):
        histTimeliness = 2.0
    elif (diffDaysMin > -22.0 and diffDaysMin <= -19.0):
        histTimeliness = 3.0
    elif (diffDaysMin > -19.0 and diffDaysMin <= -16.0):
        histTimeliness = 4.0
    elif (diffDaysMin > -16.0 and diffDaysMin <= -13.0):
        histTimeliness = 5.0
    elif (diffDaysMin > -13.0 and diffDaysMin <= -10.0):
        histTimeliness = 6.0
    elif (diffDaysMin > -10.0 and diffDaysMin <= -7.0):
        histTimeliness = 7.0
    elif (diffDaysMin > -7.0 and diffDaysMin <= -4.0):
        histTimeliness = 8.0
    elif (diffDaysMin > -4.0 and diffDaysMin < -1.0):
        histTimeliness = 9.0
    else:
        histTimeliness = 10.0

    return histTimeliness
```

```python
import math

for i in range(len(allSuppliers)):
    supplierId = allSuppliers[i]
    for j in range(len(allCommodities)):
        commodityId = allCommodities[j]
        
        # Compute average quality and timeliness
        supplier_item = supplierCharacteristicsDf.query('SupplierId == "'+str(supplierId)+ '" & CommodityId == '+str(commodityId))
        print(supplier_item.index)
        quality = 0.0
        timeliness = 0.0
        
        # PercentageReceived	DiffDays
        if (supplier_item.size > 0):
            poID = supplier_item['POId'].unique()
            #Compute Quality
            percentageReceivedMin  = supplier_item['PercentageReceived'].min()
            percentageReceivedMax = supplier_item['PercentageReceived'].max()
            percentageReceivedMean = supplier_item['PercentageReceived'].mean()
            quality = computeHistQuality(percentageReceivedMin, percentageReceivedMax, percentageReceivedMean)
            
            #Compute Timeliness
            diffDaysMin = supplier_item['DiffDays'].min()
            diffDaysMax = supplier_item['DiffDays'].max()
            diffDaysMean = supplier_item['DiffDays'].mean()
            timeliness = computeHistTimeliness(diffDaysMin, diffDaysMax, diffDaysMean)
            
            print("Quality=" +str(quality))
            print("Timeliness=" +str(timeliness))
            updateIndices = supplier_item.index
            for k in range(len(updateIndices)):
                supplierCharacteristicsDf.at[updateIndices[k],'HistQuality'] = quality
                supplierCharacteristicsDf.at[updateIndices[k],'HistTimeliness'] = timeliness

supplierCharacteristicsDf.to_csv('./output/supplier_characteristics_details.csv', mode='w', header=True, encoding='utf-8', index=False)
supplierCharacteristicsDf
```
